# EDA stands for Exploratory Data Analysis.         :


It is the process of exploring, understanding, and analyzing data before building a machine learning model.

# The main goal is to answer questions like:

What does the data look like?

Are there missing values?

Are there outliers?

Which features are important?
:
Are patterns or relationships present?

# **Simple Definition**

# EDA is like “getting to know the dataset” before using it for machine learning.

Just like a doctor examines a patient before treatment, a data analyst examines data before modeling.

| EDA Task                | Purpose                        |
| ----------------------- | ------------------------------ |
| Viewing data            | Understand structure           |
| Checking missing values | Find incomplete data           |

Understanding Each Operation

Creates a duplicate dataset.

Without this, the original dataset changes permanently.

mean()

Calculates average.

---



Mean formula:

xˉ=∑x/n

Where:

∑x = sum of all values

n = number of values




Example:

Ages = 20, 30, 40

Mean: (20+30+40)/3=30

fillna(mean_age)

Replaces missing values using the calculated average.

**Method 3: Fill Missing Categorical Values Using Mode**
mode_town = titanic['embark_town'].mode()[0]

titanic['embark_town'] = titanic['embark_town'].fillna(mode_town)

**Understanding Mode**

Mode = most frequent value.

Example:

City
Chennai
Bangalore
Chennai

Mode = Chennai

```python
titanic_mean = titanic.copy()

mean_age = titanic_mean['age'].mean()

titanic_mean['age'] = titanic_mean['age'].fillna(mean_age)

print("Missing Values After Filling:")
print(titanic_mean['age'].isnull().sum())
```

Outlier Detection
What is an Outlier?

An outlier is an unusually different value.

Example:

Ages
20
25
23
200

200 is suspicious.

```python
Q1 = titanic['age'].quantile(0.25)
Q3 = titanic['age'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = titanic[(titanic['age'] < lower_bound) |
                   (titanic['age'] > upper_bound)]

print(outliers[['age']].head())
```

4. Feature Scaling

Including:

Min Max Scaling
Standardization

is very useful because many tutorials skip preprocessing explanations.

Students learn why scaling matters for algorithms like:

KNN
SVM
Neural Networks

we make all columns between -1 - 1 so that they do not reduce the importance of other columns just being larger in magnitude does not mean more impact.

Scaling is not done to improve the data. Scaling is done to prevent large-number features from unfairly dominating algorithms that depend on distances or numerical magnitude.
1. Min-Max Scaling (Range 0-1) (Normalisation)
2. Standarisation (Z Score sclaing) - use mean and std dev. No fixed range

Transformation - when data is very much skewed. changes the shape of the distribution, not just the scale. Reduce skewness. Make data more normal
Log Transform
Square Root Transform
Box-Cox Transform
Yeo-Johnson Transform

Standardization - Shape stays same. make mean = 0. std = 1
Everything measured in "how far from average. makes data equally distributed

Variance
=
Average squared distance from mean

Standard Deviation
=
Square root of variance
